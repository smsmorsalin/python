problem_07: Take n as input and print the pyramid shape of *

Example of output


```
Enter how many rows will be in the pyramid: 5

  * * *

* * * * *
```




````python
#method_01 (using while loop)
n = int(input('Enter how many rows will be in the pyramid: '))
i = 1
while i <= n:
  j = 1
  while j <= n-i:
    print(' ',end="")
    j+=1
  k = 1
  while k <= i:
    print('*',end=' ')
    k+=1
  print()
  i+=1
````

````python
#method_02 (using for loop)
n = int(input('Enter the number of line of the pyramid: '))
for i in range(n):
  for j in range(n-i):
    print('',end=' ')
  for k in range(i):
    print('*',end=' ')
  print()
````

Then prints the principal amount, interest amount, total payable and customer id.

````python
def calculate_and_pay_interest(account_balance, interest_rate, customer_id):
    # Ensure minimum interest rate is 0.5%
    if interest_rate < 0.5:
        interest_rate = 0.5

    # Calculate the interest amount
    interest_amount = (account_balance * interest_rate) / 100

    # Calculate the total payable (principal + interest)
    total_payable = account_balance + interest_amount

    # Print the results
    print(f"Customer ID: {customer_id}")
    print(f"Principal Amount: ${account_balance:.2f}")
    print(f"Interest Amount: ${interest_amount:.2f}")
    print(f"Total Payable: ${total_payable:.2f}")

# Example usage of the function
account_balance = float(input("Enter account balance: $"))
interest_rate = float(input("Enter interest rate (as percentage): "))
customer_id = input("Enter customer ID: ")

# Call the function to calculate interest and print details
calculate_and_pay_interest(account_balance, interest_rate, customer_id)
````
