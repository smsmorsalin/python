````python
#for loop

#Method_01 (start, stop, step)
#start: 1
#stop: n+1
#step: +1
n = int(input('Enter a positive integer number: '))
for count in range (1, n+1, 1):
  print(count)

print('\n Method_02 .....................................')

#method_02 (start, stop)
#start: 1
#stop: n+1
#step: +1 (default)
for count in range (1, n+1) :
  print(count)

print('\n Method_03 .....................................')

#method_03 (stop)
#start: 0 (default)
#stop: n+1
#step: +1 (default)
for count in range (n+1):
  if count == 0:
    continue          #skip 0 (learn from w3school)
  else:
    print(count)
````
